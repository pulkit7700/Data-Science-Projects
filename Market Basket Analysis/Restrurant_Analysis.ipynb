```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df
```

```python
d1 = df.drop(["id", "total"], axis=1)
```

```python
d1
```

```python
transform = d1.fillna("!")
transform
```

```python
trans = transform.values.tolist()
```

```python
for i in range(len(trans)):
  trans[i] = [x for x in trans[i] if not x=="!"]
```

```python
trans 
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_array = te.fit_transform(trans)
transactions = pd.DataFrame(te_array, columns=te.columns_)
```

```python
freq_items = apriori(transactions, min_support=0.001, use_colnames=True)
freq_items
```

```python
freq_items["length"] = freq_items["itemsets"].apply(lambda x: len(x))
```

```python
freq_items = freq_items.sort_values('length', ascending=False)
```

```python
freq_items
```

```python
rules = association_rules(freq_items, metric='confidence', min_threshold=0.001)
rules
```

```python
final = rules.sort_values('confidence', ascending=False)
final
```

```python
final.to_excel("final.xlsx")
```
